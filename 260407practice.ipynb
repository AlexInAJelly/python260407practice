```python
print("計算最終成績")
score = input("請輸入分數:")
sc = int(score)

if(sc>=90):
    print("A")
elif(sc>=80):
    print("B")
elif(sc>=70):
    print("C")
elif(sc>=60):
    print("D")
else:
    print("F")
```

```python
print("計算生肖")
yyyy = input("請輸入西元出生年:")
y = int(yyyy) - 1900

if(y%12 == 0):
    print("屬老鼠")
elif(y%12 == 1):
    print("屬牛")
elif(y%12 == 2):
    print("屬老虎")
elif(y%12 == 3):
    print("屬兔")
elif(y%12 == 4):
    print("屬龍")
elif(y%12 == 5):
    print("屬蛇")
elif(y%12 == 6):
    print("屬馬")
elif(y%12 == 7):
    print("屬羊")
elif(y%12 == 8):
    print("屬猴")
elif(y%12 == 9):
    print("屬雞")
elif(y%12 == 10):
    print("屬狗")
else:
    print("屬豬")
```

```python
print("計算票價")
age = input("請輸入年齡:")
age = int(age)
ticket = 100

if (age<=6 or age>=80):
    ticket = ticket * 0.2
    print("票價是%d元"%ticket)
elif (age<=12 and age>=7):
    ticket = ticket * 0.5
    print("票價是%d元"%ticket)
elif (age<=79 and age>=60):
    ticket = ticket * 0.5
    print("票價是%d元"%ticket)
else:
    print("票價是%d元"%ticket)
```
